```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the cleaned dataset
df_cleaned = pd.read_csv('cleaned_vgsales.csv')
```

```python
df_cleaned.head()
```

##### Gaming genre with the highest total global sales

```python
# Group the data by "Genre" and calculate total global sales
genre_sales = df_cleaned.groupby('Genre')['Global_Sales'].sum()

# Find the genre with the highest total global sales
genre_with_highest_sales = genre_sales.idxmax()
highest_sales = genre_sales.max()

print(f"The gaming genre with the highest total global sales is '{genre_with_highest_sales}' with total sales of {highest_sales} million.")
```

##### Trends in the popularity of genres over the years

```python
# Group the data by "Year" and "Genre" and calculate total global sales
genre_sales_by_year = df_cleaned.groupby(['Year', 'Genre'])['Global_Sales'].sum().unstack()

# Create a line plot to visualize genre trends over time
plt.figure(figsize=(12, 6))
plt.title('Gaming Genre Popularity Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Global Sales (in millions)')
genre_sales_by_year.plot(ax=plt.gca())
plt.legend(loc='upper left', bbox_to_anchor=(1.05, 1))

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Create a line plot using Seaborn to visualize genre trends over time
plt.figure(figsize=(12, 6))
plt.title('Gaming Genre Popularity Over the Years (Seaborn)')
plt.xlabel('Year')
plt.ylabel('Total Global Sales (in millions)')
sns.lineplot(data=genre_sales_by_year, dashes=False)

# Show the plot
plt.tight_layout()
plt.show()
```

##### Genre with the highest average sales per game

```python
# Group the data by "Genre" and calculate average global sales per game
genre_avg_sales = df_cleaned.groupby('Genre')['Global_Sales'].mean()

# Find the genre with the highest average sales per game
genre_with_highest_avg_sales = genre_avg_sales.idxmax()
highest_avg_sales = genre_avg_sales.max()

print(f"The gaming genre with the highest average sales per game is '{genre_with_highest_avg_sales}' with an average sales of {highest_avg_sales:.2f} million.")
```
